# 1. Paint could never be more fun

Acest task va consta intr-un mic program ce va manipula cursorul astfel incat, avand deschis un program precum Paint, vom incerca sa desenam un shooting star si sa ne scriem initialele pe opera de arta rezultata.

Pasii pe care ii vom urma sunt:

```python
# TO DO 1.1: Importam modulele de care avem nevoie: pyautogui si time -- de ce time, explicatie mai jos.
import pyautogui
import time

# TO DO 1.2: Ne asiguram ca putem opri executia programului oricand, folosind feature-ul FAILSAFE, si, daca este
# nevoie setam un delay intre instructiuni folosind PAUSE
pyautogui.PAUSE = 0.3 
pyautogui.FAILSAFE = True

# Daca am rula aceasta celula pur si simplu, instructiunile se vor rula imediat, fara ca noi sa avem ocazia de a 
# deschide paint-ul inainte. Astfel, avem nevoie de putin timp pentru a plasa cursorul peste canvas-ul din Paint
# inainte ca el sa inceapa sa deseneze.

# TO DO 1.3 Setam un delay folosind functia sleep() din modulul time, careia ii dam ca parametrul numarul de 
# secunde de delay
time.sleep(7)

# TO DO 1.4 Vrem sa ne asiguram ca fereastra Paint se afla in focus (este selectata) asa ca, just in case, vom
# efectua un click pe ea
pyautogui.click()

# TO DO 1.5 Folosind functia dragRel() sau dragTo(), definiti o functie care sa deseneze o stea.
# Mai tarziu vom vrea sa ilustram o stea cazatoare, asa ca ar fi bine ca functiei drawStar() sa ii dam si un 
# parametru prin care sa scalam stelele desenate
def drawStar(param):
    # Se deseneaza liniile ce formeaza stelele. 
    # NOTE: Cu cat desenam mai multe stele, cu atat mai mari am vrea ca ele sa fie, pentru a da efectul de
    # apropiere, deci pentru a obtine acest lucru se inmultesc coordonatele cu factorul de scalare primit ca parametru.
    pyautogui.dragRel(20 * param, -30 * param, duration = 0.2)
    pyautogui.dragRel(20 * param, 30 * param, duration = 0.2)
    pyautogui.dragRel(-45 * param, -17 * param, duration = 0.2)
    pyautogui.dragRel(50 * param, 0, duration = 0.2)
    pyautogui.dragRel(-45 * param, 17 * param, duration = 0.2)

# TO DO 1.6 Se definesc factorul de scalare si numarul de stele pe care vrem sa il desenam
enlarge = 1
repeat = 3

# TO DO 1.7 Folosim un loop pentru a desena atatea stele cate am definit mai sus
while repeat > 0:
    # Se deseneaza steaua
    drawStar(enlarge)
    # Se muta cursorul (mai jos, putin spre dreapta poate?)
    pyautogui.moveRel(200, -50)
    # Creste factorul de scalare
    enlarge = enlarge*2
    # Scade numarul de stele care mai trebuie desenate
    repeat = repeat - 1;
    
time.sleep(1)
# TO DO 1.8 Folosind dragRel() sau dragTo() desenati-va initialele pe aceeasi schita din Paint.
# De exemplu, pentru a desena litera d avem nevoie de urmatoarele comenzi:
# pyautogui.moveTo(240, 200)
# pyautogui.dragTo(190, 200, duration = 0.2)
# pyautogui.dragTo(190, 250, duration = 0.2)
# pyautogui.dragTo(240, 250, duration = 0.2)
# pyautogui.dragTo(240, 150, duration = 0.2)

#c
pyautogui.moveTo(100, 200)
pyautogui.dragTo(50, 200, duration = 0.2)
pyautogui.dragTo(50, 250, duration = 0.2)
pyautogui.dragTo(100, 250, duration = 0.2)
#o
pyautogui.moveTo(120, 200)
pyautogui.dragTo(170, 200, duration = 0.2)
pyautogui.dragTo(170, 250, duration = 0.2)
pyautogui.dragTo(120, 250, duration = 0.2)
pyautogui.dragTo(120, 200, duration = 0.2)
#d
pyautogui.moveTo(240, 200)
pyautogui.dragTo(190, 200, duration = 0.2)
pyautogui.dragTo(190, 250, duration = 0.2)
pyautogui.dragTo(240, 250, duration = 0.2)
pyautogui.dragTo(240, 150, duration = 0.2)
#e
pyautogui.moveTo(310, 250)
pyautogui.dragTo(260, 250, duration = 0.2)
pyautogui.dragTo(260, 200, duration = 0.2)
pyautogui.dragTo(310, 200, duration = 0.2)
pyautogui.dragTo(310, 225, duration = 0.2)
pyautogui.dragTo(260, 225, duration = 0.2)
#t
pyautogui.moveTo(355, 150)
pyautogui.dragTo(355, 250, duration = 0.2)
pyautogui.moveTo(330, 200)
pyautogui.dragTo(380, 200, duration = 0.2)
#t
pyautogui.moveTo(425, 150)
pyautogui.dragTo(425, 250, duration = 0.2)
pyautogui.moveTo(400, 200)
pyautogui.dragTo(450, 200, duration = 0.2)
#e
pyautogui.moveTo(520, 250)
pyautogui.dragTo(470, 250, duration = 0.2)
pyautogui.dragTo(470, 200, duration = 0.2)
pyautogui.dragTo(520, 200, duration = 0.2)
pyautogui.dragTo(520, 225, duration = 0.2)
pyautogui.dragTo(470, 225, duration = 0.2)
```

# 2. Download a random image from the internet

Acest task va consta intr-un program care sa deschida un browser, sa caute imagini dupa un cuvant specificat de voi, sa salveze una din imagini pe desktop, dupa care sa inchida browserul deschis.

```python
# TO DO 2.1: Importam modulele de care avem nevoie: pyautogui time si random
import pyautogui
import time
import random

# TO DO 2.2: Ne asiguram ca putem opri executia programului oricand, folosind feature-ul FAILSAFE, si, daca este
# nevoie setam un delay intre instructiuni folosind PAUSE
pyautogui.PAUSE = 0.3 
pyautogui.FAILSAFE = True

# TO DO 2.3 Open up Chrome: apasam butonul windows, tastam Chrome si apasam enter. Asteptam 0.5s sa se incarce pagina
pyautogui.press('win')
pyautogui.typewrite('chrome', 0.25)
pyautogui.press('enter')
time.sleep(0.5)

# TO DO 2.4 Cautam imagini cu [chestie]: scriem [chestie] images, apasam enter, asteptam 2s sa se incarce pagina
pyautogui.typewrite('python images', 0.15)
pyautogui.press('enter')
time.sleep(2)

# TO DO 2.5 Intram in tabul images: deschidem filtrul, cautam entry-urile cu numele 'images for [chestie]', apasam enter
pyautogui.hotkey('ctrl', 'f')
pyautogui.typewrite('images for python', 0.15)
pyautogui.hotkey('ctrl', 'enter')

# TO DO 2.6: Alegem coordonate random: hint random.randrange, verificam ca nu cumva cursorul sa ajunga intr-un spatiu
# de culoare alba (spatiul dintre poze). Atata timp cat pointer-ul nu este pe o poza, se aleg alte coordonate random
xCoord = random.randrange(57, 1550)
yCoord = random.randrange(289, 812)
while pyautogui.pixelMatchesColor(xCoord, yCoord, (255, 255, 255)):
    xCoord = random.randrange(57, 1550)
    yCoord = random.randrange(289, 812)

# TO DO 2.7: Save as: mutam cursorul la coordonatele random, click dreapta, navigam cu sagetile la Save as, apoi enter     
pyautogui.moveTo(xCoord, yCoord)
time.sleep(0.5)
pyautogui.click()
time.sleep(1)
pyautogui.moveTo(1177, 439)
pyautogui.click(button = "left")
time.sleep(1)
pyautogui.hotkey('ctrl', 's')
time.sleep(1)

# TO DO 2.8: Finalizare: ii dam imaginii un nume, navigam la calea de fisier si alegem Desktop, navigam la Save, apoi enter
pyautogui.typewrite('snek', 0.15)
for i in range(0, 6):
    pyautogui.press('tab')
for i in range(0, 2):
    pyautogui.press('down')
pyautogui.press('enter')
for i in range(0, 8):
    pyautogui.press('tab')
pyautogui.press('enter')

# TO DO 2.9 Cleaning up: Asteptam 2s sa se finalizeze descarcarea si apoi inchidem fereastra
time.sleep(2)
pyautogui.hotkey('alt', 'f4')
```

# 3. Play the T-Rex Game.

Acest task va consta intr-un mic bot care joaca T-Rex Game. We'll show 'em who's boss.

Pe scurt, programul va face urmatoarele lucruri:
 * se va pregati pentru rulare (tot ce inseamna import-uri si pause/failsafe)
 * odata aflat pe tab-ul nou cu jocul trex, va analiza imaginea si isi va procura culorile background-ului si obstacolelor
 * va incepe jocul prin apasarea tastei space
 * va analiza culoarea unui punct critic definit de noi pe tot parcursul jocului si va apasa "space" in momentul in care acest punct critic isi schimba culoarea.

```python
####### INITIALIZARE PROGRAM ########

# TO DO 1.1: Importam modulele de care avem nevoie: pyautogui, deocamdata
import pyautogui
# TO DO 1.4: Deoarece la pasul 1.3 am avut nevoie de modulul time, trebuie sa il si importam
import time

# TO DO 1.2: Ne asiguram ca putem opri executia programului oricand, folosind feature-ul FAILSAFE, si, daca este
# nevoie setam un delay intre instructiuni folosind PAUSE.
pyautogui.PAUSE = 0.1
pyautogui.FAILSAFE = True

# TO DO 1.3: Apelam functia sleep din modulul time pentru aproximativ 5 secunde, pentru a avea timp sa ne mutam de pe un
# tab / window pe altul.
time.sleep(5)




######## ANALIZARE INITIALA A IMAGINII #############

# TO DO 2.1: Se efectueaza un screenshot si se salveaza intr-un obiect de tip imagine. Nu este necesara salvarea imaginii
# in calculator.
image = pyautogui.screenshot()

# Vrem sa salvam doua culori diferite. Una pentru background si una pentru obstacole. Luckily for us, obstacolele
# au aceeasi culoare ca si dinozaurul, asa ca putem analiza chiar dinozaurul ca sa aflam culoarea obstacolelor.

# TO DO 2.2: Ne definim coordonatele a doua puncte de pe imagine. Un punct despre care stim sigur ca este un pixel
# de fundal, si punct ce reprezinta un pixel care se afla in interiorul dinozaurului. Vom folosi aceste coordonate
# ca sa aflam culorile.
backgroundPixel = (250, 250)
dinoPixel = (519, 223)

# TO DO 2.3: Folosind obiectul imagine creat anterior si coordonatele definite mai sus, salvam in doua variabile
# culoarea de background si culoarea obstacolelor.
backgroundColor = image.getpixel(backgroundPixel)
stuffsColor = image.getpixel(dinoPixel)

# TO DO 2.4: Ne definim coordonatele unui punct critic. Un punct in care, daca se inregistreaza un obstacol, vom
# apasa tasta space pentru ca dinozaurul sa sara.
criticalPoint = (615, 215)



######## INCEPERE JOC #########

# TO DO 3.1: Se apasa tasta space pentru a incepe jocul
pyautogui.press('space')

# (Optional) TO DO 3.2: Se muta cursorul la coordonatele punctului critic pentru o mai buna monitorizare a obstacolelor
# care trec pe acolo
pyautogui.moveTo(criticalPoint)



######## RULAREA EFECTIVA A PROGRAMULUI CARE JOACA T-REX GAME ############
# Se va efectua cate un sceenshot la fiecare moment de timp si se va analiza culoarea punctului critic.
# Daca acesta are culoarea dinozaurului, se va apasa tasta space.

# TO DO 4.6: Tot ce am scris in aceasta sectiune trebuie sa se afle intr-un bloc while. Vrem sa rulam programul cat de
# putem.
while(True):
    
    # TO DO 4.4: Se observa la rulare ca avem de aface cu o exceptie Keyboard Interrupt. O manevram cu un try - except
    # block. Tot ce am scris pana acum se introduce in blocul try
    try: 
        # TO DO 4.1: Se efectueaza screenshot si se salveaza obiectul imagine intr-o variabila.
        im = pyautogui.screenshot()
        
        # TO DO 4.2: Se salveaza culoarea pixelului de la punctul critic.
        pixel = im.getpixel(criticalPoint)
        print(pixel)  
        
        # TO DO 4.3: Se face verificarea culorii aflate mai sus si se apasa space in cazul corespunzator.
        if(pixel != backgroundColor):
            pyautogui.press('space')
    
    # TO DO 4.5: In cazul in care prindem exceptia, o printam, for debugging purposes si apoi apelam continue
    # ca sa ne asiguram ca programul ruleaza in continuare
    except KeyboardInterrupt:
        print('Keyboard interrupt')
        continue
```
